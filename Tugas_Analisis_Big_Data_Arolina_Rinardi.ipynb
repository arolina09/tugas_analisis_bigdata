# 1. Student Assassment

```python
SA
```

**SA** merupakan data dari file csv yang bernama Student Assassment,lalu dipanggil dan memunculkan tampilan data didalam file csv tersebut.

```python
 SA.describe()
```

**describe()** merupakan sebuah method yang dimiliki oleh python di Pandas digunakan untuk menghitung statistik dasar untuk kolom numerik dan string di DataFrame. Jadi jika kita menginginkan menghitung data statistika dari data student assassment maka dengan perintah memanggil SA.descibe().

```python
SA.dtypes
```

**dtypes** adalah atribut yang tersedia pada objek DataFrame dalam perpustakaan Pandas di Python. Atribut ini digunakan untuk mengembalikan tipe data dari setiap kolom dalam DataFrame.

```python
# Menampilkan metadata dataset
SA.info()
```

**Metode info()** di Pandas digunakan untuk mendapatkan ringkasan singkat dari DataFrame, termasuk informasi tentang jumlah baris, jumlah nilai bukan nol di setiap kolom, dan tipe data setiap kolom.

```python
# Mengecek komponen-komponen dataset
print(f"Data Types:\n{SA.dtypes}")
print(f"\nRows and Columns:\n{SA.shape}")
print(f"\nColumn Names:\n{SA.columns}")
print(f"\nNull Values %:\n{SA.apply(lambda x: sum(x.isnull()) / len(SA))}")
```

Kode diatas memberikan informasi penting tentang DataFrame SA, termasuk tipe data kolom, ukuran DataFrame, daftar nama kolom, dan persentase nilai null dalam setiap kolom. Informasi ini bermanfaat untuk memahami struktur dan isi DataFrame dan dapat membantu dalam analisis dan pemrosesan data lebih lanjut.

```python
#melihat apakah di data ini memiliki missing value?
SA.isnull().sum()
```

Kode SA.isnull().sum()menghitung jumlah nilai yang hilang di setiap kolom dan mengembalikan objek Seri dengan nama kolom sebagai indeks dan jumlah nilai yang hilang sebagai nilainya. jadi dalam data ini memiliki missing value pada fitur score sejumlah 173

```python
# Data Cleaning analisis yang baik
rata = SA['score'].mean()
SA['score'] = SA['score'].fillna(rata)
SA['score'].isna().sum()
```

mengatasi masalah nilai null dengan menggantinya dengan nilai yang masuk akal (rata-rata dalam hal ini). Dengan melakukan hal ini, kita dapat menjaga integritas data dan memastikan bahwa kolom 'score' tidak memiliki nilai null yang dapat mengganggu analisis atau pemrosesan data selanjutnya.

```python
SA.isnull().sum()
```

melihat kembali apakah setelah dilakukan menghilangkan missing value pada fitur score.
